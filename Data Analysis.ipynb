```python
# Ques#1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Creating the DataFrame
data = {
    "ID": range(1, 11),
    "Age": [28, 35, 42, 30, 50, 41, 29, 33, 45, 39],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
    "Department": ["IT", "HR", "Finance", "Marketing", "Sales", "HR", "IT", "Finance", "Sales", "Marketing"],
    "Salary ($)": [70000, 60000, 90000, 65000, 85000, 62000, 75000, 88000, 87000, 67000],
    "Experience (Years)": [3, 7, 12, 5, 15, 9, 4, 11, 14, 6],
    "Work Hours/Week": [40, 38, 45, 42, 50, 37, 39, 44, 48, 41],
    "Projects Completed": [5, 4, 6, 7, 8, 5, 6, 7, 9, 6],
    "Job Satisfaction (1-10)": [7, 8, 6, 9, 5, 7, 8, 6, 5, 8],
    "Training Hours": [20, 15, 18, 25, 12, 16, 22, 20, 14, 19],
    "Promotion Status": ["No", "Yes", "Yes", "No", "Yes", "No", "No", "Yes", "Yes", "No"]
}

df = pd.DataFrame(data)




```

```python

# i) Determine total number of rows and columns
rows, cols = df.shape
print(f"Total Rows: {rows}, Total Columns: {cols}")

```

```python
# ii) Summary statistics for numerical attributes
print(df.describe())

```

```python
# iii) Count distinct values in Department and Promotion Status
print(df["Department"].value_counts())
print(df["Promotion Status"].value_counts())

```

```python
# Handling Missing Values
# i) Checking for missing values
print(df.isnull().sum())
```

```python
# ii) Handling missing values using mean, median, or mode
df.fillna({
    "Age": df["Age"].mean(),
    "Salary ($)": df["Salary ($)"].median(),
    "Experience (Years)": df["Experience (Years)"].mean(),
    "Work Hours/Week": df["Work Hours/Week"].mean(),
    "Projects Completed": df["Projects Completed"].median(),
    "Job Satisfaction (1-10)": df["Job Satisfaction (1-10)"].median(),
    "Training Hours": df["Training Hours"].mean(),
    "Promotion Status": df["Promotion Status"].mode()[0]
}, inplace=True)
```

Visual Analysis

```python
# Visual Analysis
# i) Histogram of Salaries
plt.figure(figsize=(8, 5))
sns.histplot(df["Salary ($)"], bins=5, color='pink')
plt.title("Distribution of Employee Salaries")
plt.xlabel("Salary ($)")
plt.ylabel("Frequency")
plt.show()

```

```python
# ii) Bar graph for number of employees per department
plt.figure(figsize=(8, 5))
sns.countplot(x=df["Department"], palette="viridis", hue=None)
plt.title("Number of Employees in Each Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()

```

```python
# iii) Box plot for Experience (Years)
plt.figure(figsize=(6, 5))
sns.boxplot(y=df["Experience (Years)"], color='green')
plt.title("Box Plot of Employee Experience (Years)")
plt.ylabel("Experience (Years)")
plt.show()
```

```python
#Ques#2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the Heart Disease dataset
df = pd.read_csv("heart_disease.csv")

```

```python
# Basic Insights
# i) Determine total number of rows and columns
rows, cols = df.shape
print(f"Total Rows: {rows}, Total Columns: {cols}")
```

```python
# ii) Summary statistics for numerical attributes
print(df.describe())
```

```python
# iii) Count distinct values in Gender and Smoking columns
print(df["Gender"].value_counts())
print(df["Smoking"].value_counts())
```

```python
# Handling Missing Data
# i) Checking for missing values
print(df.isnull().sum())
```

```python
# ii) Handling missing values
# Mean for continuous numerical data, mode for categorical data
df.fillna({
    "Age": df["Age"].mean(),
    "Cholesterol": df["Cholesterol Level"].median(),
    "Gender": df["Gender"].mode()[0],
    "Smoking": df["Smoking"].mode()[0]
}, inplace=True)
```

Data visualization

```python
# Data Visualization
# i) Histogram of Age
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=10, color='blue')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# ii) Bar chart for Heart Disease Status
plt.figure(figsize=(6,4))
sns.countplot(x=df['Heart Disease Status'], palette="viridis")
plt.title("Count of People with and without Heart Disease")
plt.xlabel("Heart Disease Status")
plt.ylabel("Count")
plt.show()
```

```python

# iii) Box plot for Cholesterol Level
plt.figure(figsize=(6, 5))
sns.boxplot(y=df["Cholesterol Level"], color='green')
plt.title("Box Plot of Cholesterol Levels")
plt.ylabel("Cholesterol Level")
plt.show()
```
